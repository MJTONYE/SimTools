Mean of π̂

Der Mittelwert der 20 Replikationen nähert sich mit wachsendem N immer stärker an den wahren Wert π ≈ 3.14159 an.
Empirical SE across replications

Die empirische Streuung der π̂-Schätzer über 20 Replikationen skaliert ungefähr wie 1/√N, wie in der Theorie erwartet.

95%-CI for the mean

Das 95%-Konfidenzintervall für den Mittelwert wird für große N sehr eng und enthält durchgängig den wahren Wert π.

Interpretation der 1/√N-Skalierung

Die Streuung der π̂-Schätzungen nimmt ungefähr proportional zu 1/√N ab,
wie es die Monte-Carlo-Theorie vorhersagt.
Verdoppelt man z. B. die Stichprobengröße, halbiert sich der Standardfehler nicht,
aber er fällt um den Faktor 1/√2.

Dies zeigt, dass die Monte-Carlo-Methode relativ langsam konvergiert:
Um die Präzision um den Faktor 10 zu verbessern, muss N um den Faktor 100 wachsen.
