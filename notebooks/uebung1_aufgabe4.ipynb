```python
import sys, os
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, project_root)

from src.integration import g1, g2, g3, mc_expectation, uniform_sampler
import numpy as np
import time

rng = np.random.default_rng(42)

Ns = [1_000, 10_000, 100_000]
functions = {"g1": g1, "g2": g2, "g3": g3}

results = []

for name, g in functions.items():
    for N in Ns:
        start = time.perf_counter()
        mu_hat, se = mc_expectation(g, uniform_sampler, N, rng)
        duration = time.perf_counter() - start
        
        results.append([name, N, mu_hat, se, duration])

results

import pandas as pd

df = pd.DataFrame(results, columns=["g", "N", "mu_hat", "SE", "runtime_sec"])
df
```
